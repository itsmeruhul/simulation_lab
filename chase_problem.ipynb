```python
import math
def dis(x1,y1,x2,y2):
  return math.sqrt((x1-x2)**2 + (y1-y2)**2)
```

```python
x_A=[10]
y_A=[0]
x_B=[0]
y_B=[10]
x_C=[10]
y_C=[10]
x_D=[0]
y_D=[0]

v_A=3
v_B=5
v_C=7
v_D=2

D_got_shoot = 0
B_got_shoot = 0
C_got_shoot = 0
A_got_shoot = 0



distDB= dis(x_D[0],y_D[0],x_B[0],y_B[0])
distCA= dis(x_C[0],y_C[0],x_A[0],y_A[0])
distBC= dis(x_B[0],y_B[0],x_C[0],y_C[0])
distAD= dis(x_A[0],y_A[0],x_D[0],y_D[0])

for t in range(1,21):
  #D is chased by B target-D 
  sin = (y_D[t-1]- y_B[t-1]) / distDB
  cos= (x_D[t-1] - x_B[t-1]) / distDB
  new_x_A= x_B[t-1]+ v_B* cos
  new_y_A= y_B[t-1] + v_B*sin
  x_B.append(new_x_B)
  y_B.append(new_y_B)

  #C is chased by A Target- C 
  sin = (y_C[t-1]- y_A[t-1]) / distCA
  cos= (x_C[t-1] - x_A[t-1]) / distCA
  new_x_A= x_A[t-1]+ v_A* cos
  new_y_A= y_A[t-1] + v_A*sin
  x_A.append(new_x_A)
  y_A.append(new_y_A)

  #B is chased by C
  sin = (y_B[t-1] - y_C[t-1]) / distBC
  cos = (x_B[t-1] - x_C[t-1]) / distBC
  new_x_C =  x_C[t-1] + v_C * cos
  new_y_C =  y_C[t-1] + v_C * sin
  x_C.append(new_x_C)
  y_C.append(new_y_C)


  # A is chased by D

  sin = (y_A[t-1] - y_D[t-1]) / distAD
  cos = (x_A[t-1] - x_D[t-1]) / distAD
  new_x_D =  x_D[t-1] + v_D * cos
  new_y_D =  y_D[t-1] + v_D * sin
  x_D.append(new_x_D)
  y_D.append(new_y_D)



  


  distDB= dis(x_D[t],y_D[t],x_B[t],y_B[t])
  distCA= dis(x_C[t],y_C[t],x_A[t],y_A[t])
  distBC= dis(x_B[t],y_B[t],x_C[t],y_C[t])
  distAD= dis(x_A[t],y_A[t],x_D[t],y_D[t])





  if distDB<5:
    print("t = ",t)
    print("distance_of_DB = " , distDB)
    print("B shoot D")
    D_got_shoot = D_got_shoot + 1

  if distAD<5:
    print("t = ",t)
    print("distance_of_AD = " , distAD)
    print("D shoot A")
    A_got_shoot = A_got_shoot + 1  

  if distCA<5:
    print("t = ",t)
    print("distance_of_CA = " , distCA)
    print("A shoot C")
    C_got_shoot = C_got_shoot + 1

  if distBC<5:
    print("t = ",t)
    print("distance_of_BC = " , distBC)
    print("C shoot B")
    B_got_shoot = B_got_shoot + 1

   
 





print("A_got_shoot = ",A_got_shoot)
print("B_got_shoot = ",B_got_shoot)

print("C_got_shoot = ",C_got_shoot)
print("D_got_shoot = ",D_got_shoot)
```

```python
import matplotlib.pyplot as plt
plt.plot( x_A,y_A, color='yellow', linewidth=2, label ='A')
plt.plot( x_B,y_B, color='black', linewidth=2, label ='B')
plt.plot( x_C,y_C, color='blue', linewidth=2, label ='C')
plt.plot( x_D,y_D, color='green', linewidth=2, label ='D')
plt.legend()
```
