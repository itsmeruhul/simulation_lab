```
import numpy as np
import matplotlib.pyplot as plt; plt.rcdefaults()
import matplotlib.pyplot as plt
```

```
m = 20   # maximum inventory
n = 5   # review period
cycle = 10
loop = n*cycle

beginning_inventory = [3]
order_quentity = 8
days_until_order_arrives = 1
ending_inventory = []
shortage_quantity = []
day = []

a1 = [0,1,2,3,4]   # for demand
a2 = [1,2,3]       # for lead time


demand = np.random.choice(a1,200,p=[0.10,0.25,0.35,0.21,0.09])   # random demand
lead_time = np.random.choice(a2,200,p=[0.60,0.30,0.10])           # random lead time

#print(demand)
#print(lead_time)



count = 0
j = 0

for i in range(loop):

  day.append(i)

  
  
  if i <= 0:
    new_ending_inventory = beginning_inventory[i] - demand[i]
  else:
    new_ending_inventory = beginning_inventory[i] - (demand[i] + shortage_quantity[i-1])



  if new_ending_inventory>=0: #no shortage
    ending_inventory.append(new_ending_inventory)
    shortage_quantity.append(0)

    if days_until_order_arrives!=1:#days when order arrives
      beginning_inventory.append(new_ending_inventory)
    else:
      
      new_beginning_inventory = new_ending_inventory + order_quentity
      beginning_inventory.append(new_beginning_inventory)
      
    
    if count%n == 0 & i != 0:
      order_quentity = m - new_ending_inventory
      days_until_order_arrives = lead_time[j] 
      j = j+1
      count = 0
      print ("order_quantity = ",order_quentity)
      print ("days_until_order_arrives = ",days_until_order_arrives)
    
  else: #shortage
    ending_inventory.append(0)
    new_shortage_quantity = abs(new_ending_inventory)
    shortage_quantity.append(new_shortage_quantity)

    if days_until_order_arrives!= 1:
      beginning_inventory.append(0)
    else:
      
      new_beginning_inventory = order_quentity
      beginning_inventory.append(new_beginning_inventory)
      



    if count%n == 0:
      order_quentity = m
      days_until_order_arrives = lead_time[j] 
      j = j+1
      count = 0

      print ("order_quantity = ",order_quentity)
      print ("days_until_order_arrives = ",days_until_order_arrives)


  
  count = count+1
  days_until_order_arrives = days_until_order_arrives - 1 

  print("beginning_inventory = ",beginning_inventory[i])
  print("demand = ",demand[i])
  print("ending_inventory = ",ending_inventory[i])
  print("shortage_quantity = ",shortage_quantity[i])
  print('\n')
  



print(shortage_quantity)

shortage_occurs = np.count_nonzero(shortage_quantity)
print("shortage_occurs = " , shortage_occurs)

average_ending_inventory = np.sum(ending_inventory)/len(ending_inventory)
print("average_ending_inventory = ",average_ending_inventory)

plt.bar(day,ending_inventory)
plt.show()
```
