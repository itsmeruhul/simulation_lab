```
import random
import math
import matplotlib.pyplot as plt
```

```
n=[100,1000,5000,10000]
hits=0
hit=[]
miss=[]
pi_1=[]
for i in range(len(n)):
  hits=0
  for j in range(n[i]):

    x=random.randint(0,1)
    y=random.randint(0,1)
    val = math.sqrt(x**2+y**2)
    if val<=3:
      hits=hits+1
      hit.append(val)
    else:
      miss.append(val)  


  PI = 4* (hits/n[i])
  pi_1.append(PI)
  print(pi_1)   

```
