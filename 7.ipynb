```
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]
toothpaste = [2500,2630,2140,3400,3600,2760]
facewash =[1500,1200,1340,1130,740,1555]
shampoo=[5200,5100,4550,5870,4560,4890]
moisturizer =[920,6100,9550,8870,7760,7490]
soap =[1200,2100,3550,1870,1560,1890]

plt.xlabel("month no.")
plt.ylabel("sales")

plt.plot( x, toothpaste, color='red', linewidth=2, label ='toothpaste')
plt.plot( x, facewash, color='black', linewidth=2, label ='facewash')
plt.plot( x, shampoo, color='yellow', linewidth=2, label ='shampoo')
plt.plot( x, moisturizer, color='green', linewidth=2, label ='moisturizer')
plt.plot( x, soap, color='orange', linewidth=2, label ='soap')
plt.legend()


```

```
plt.scatter(x, moisturizer,label="moisturizer")
plt.legend()
```

```
total=[]
for i in range(6):
  sum=toothpaste[i] + facewash[i] + shampoo[i] + moisturizer[i] + soap[i]
  total.append(sum)
print(total)
plt.bar(x, total, align='center', alpha=0.5)
plt.show()
```
