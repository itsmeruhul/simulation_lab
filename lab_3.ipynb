```
import random
import math
import matplotlib.pyplot as plt


```

```
n=[100,1000,5000,10000]
hits=0
pi_1=[]
x_hit=[]
y_hit=[]
x_miss=[]
y_miss=[]
for i in range(len(n)):

  for j in range(n[i]):
    
    x=random.randint(0,5)
    y=random.randint(0,5)
    val = math.sqrt(x**2+y**2)
    if val<=3:
      hits=hits+1
      x_hit.append(x)
      y_hit.append((y)



    if val>3:
      x_miss.append(x)
      y_miss.append(y)

  PI = 4* (hits/n[i])
  pi_1.append(PI)
  print(PI)    
   
x=["100","1000","5000","10000"]
plt.bar(x,pi_1,color="green")
plt.show()  
```
