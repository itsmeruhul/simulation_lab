```
import random 
import math
import matplotlib.pyplot as plt
```

```
n=[100,1000,5000,10000]
a=0
b=2
error_1=[]
for i in range(len(n)):
  func_sum=0
  func_sq_sum=0
  for j in range(n[i]):
    x=random.uniform(a,b)
    x_exp=math.exp(x)
    func_1=x**2*x_exp*math.log(x_exp)
    func_sum=func_sum+func_1
    func_sq=func_1**2
    func_sq_sum=func_sq_sum+func_sq
  func_br=func_sum/n[i] 
  func_sq_sum_br=func_sq_sum/n[i]  
  inte_value=func_br*(b-a) 
  error=((b-a)/(n[i])**.5)*(func_sq_sum_br-(func_br)**2)**.5
  error_1.append(error)
  print("n =",n[i])
  print("inte_value =",inte_value)
  print("error =",error)


x=["100","1000","5000","10000"]
plt.xlabel("Value of N") 
plt.ylabel("Value of Error") 
plt.bar(x,error_1, color ='maroon', width = 0.4) 
plt.show()  


 

```
