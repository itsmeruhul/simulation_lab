```
import random
import math
import matplotlib.pyplot as plt
```

```
n=[1000,5000,10000]

pi_1=[]
x_hit=[]
y_hit=[]
x_miss=[]
y_miss=[]
val_area= []
r =2
for i in range(len(n)):
  hits = 0
  for j in range(n[i]):
    x=random.uniform(2,8)
    y=random.uniform(2,8)
    val=math.sqrt((x-5)**2+(y-5)**2)
    if val<=r:
      hits=hits+1
      x_hit.append(x)
      y_hit.append(y)
    else :
      x_miss.append(x)
      y_miss.append(y)

  PI = 4* (hits/n[i])
  pi_1.append(PI)
  val_of_area = PI* (r**2)
  val_area.append(val_of_area)
  print("Trails = ",n[i])
  print("Hits =",hits)
  
  print("val_of_Area =", val_of_area)
  plt.scatter(x_hit, y_hit , c = "red",label='Inside_Circle')
  plt.scatter(x_miss, y_miss , c = "green",label='Outside_Circle')
  plt.legend()
  plt.show()      








```
