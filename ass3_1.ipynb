```
import random
import math
import matplotlib.pyplot as plt
```

```
n=[100,1000,5000,10000]

pi_1=[]
x_hit=[]
y_hit=[]
x_miss=[]
y_miss=[]
val_area= []
r = 3
for i in range(len(n)):
  hits = 0
  for j in range(n[i]):    
    x=random.uniform(0,5)
    y=random.uniform(0,5)
    val = math.sqrt(x**2+y**2)
    if val<=r:
      hits=hits+1
      x_hit.append(x)
      y_hit.append(y)
    else :
      x_miss.append(x)
      y_miss.append(y)
  PI = 4* (hits/n[i])
  pi_1.append(PI)
  val_of_area = PI* (r**2)
  val_area.append(val_of_area)
  print("n = ",n[i])
  print("val_of_PI = ",PI)
  print("val_of_area =", val_of_area)
  plt.scatter(x_hit, y_hit , c = "red",label='hit')
  plt.scatter(x_miss, y_miss , c = "blue",label='miss')
  plt.legend()
  plt.show()

x=["100","1000","5000","10000"]
plt.xlabel("Value of N") 
plt.ylabel("Value of PI") 
plt.bar(x,pi_1, color ='maroon', width = 0.4) 
plt.show()


```
