```
import random
import math
import matplotlib.pyplot as plt
```

```
n=[100,1000,5000,10000]

pi_1=[]
x_hit=[]
y_hit=[]
x_miss=[]
y_miss=[]
val_area= []
h = 4
b = 8
for i in range(len(n)):
  hits = 0
  for j in range(n[i]):    
    x=random.uniform(0,8)
    y=random.uniform(0,4)
    if x <= 4:
      val = math.sqrt(4*x)
      if y <= val:      
        hits=hits+1
        x_hit.append(x)
        y_hit.append(y)
      else :
        x_miss.append(x)
        y_miss.append(y)
        
    else:
      val = 8-x
      if y <= val:
        hits=hits+1
        x_hit.append(x)
        y_hit.append(y)
      else :
        x_miss.append(x)
        y_miss.append(y)

   

  val_of_rec = b*h
  val_of_tri = val_of_rec * (hits/n[i])
  val_area.append(val_of_tri)
  print("n = ",n[i])
  print("val_of_area =", val_of_tri)
  plt.scatter(x_hit, y_hit , c = "red",label='hit')
  plt.scatter(x_miss, y_miss , c = "blue",label='miss')
  plt.legend()
  plt.show()
```
